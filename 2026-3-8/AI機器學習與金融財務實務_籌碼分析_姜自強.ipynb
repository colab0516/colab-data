````python
# =========================================================
# S2330 台積電多資料來源整併
# =========================================================

# from google.colab import drive
# drive.mount('/content/drive')

import os
import pandas as pd
import numpy as np
from functools import reduce

# -------------------------------------------------
# 工作路徑
# -------------------------------------------------

BASE_PATH = f"{root}2026-3-8\\TEJ2026"
os.chdir(BASE_PATH)

print("Current Path:", os.getcwd())


# -------------------------------------------------
# Excel 檔案列表
# -------------------------------------------------

file_list = [
    "s2330d20260306.xlsx",
    "三大法人買賣超S2330台積電20260306.xlsx",
    "借券成交S2330台積電20260306.xlsx",
    "借券餘額S2330台積電20260306.xlsx",
    "當日沖銷S2330台積電20260306.xlsx",
    "融資融券S2330台積電20260306.xlsx"
]


# -------------------------------------------------
# 清理欄位名稱
# -------------------------------------------------

def clean_columns(df):

    df.columns = (
        df.columns
        .astype(str)
        .str.strip()
        .str.replace("\n","")
        .str.replace(" ","")
    )

    return df


# -------------------------------------------------
# 標準化日期欄位
# -------------------------------------------------

def standardize_date_column(df):

    candidates = ['年月日','日期','年月']

    for col in candidates:

        if col in df.columns:

            df = df.rename(columns={col:'年月日'})

            df['年月日'] = pd.to_datetime(df['年月日'], errors='coerce')

            return df

    raise ValueError("找不到日期欄位")


# -------------------------------------------------
# 讀取 Excel
# -------------------------------------------------

def read_excel_file(file):

    print("Reading:", file)

    # ⭐ 關鍵：skip 前兩列
    df = pd.read_excel(file, skiprows=2)

    df = clean_columns(df)

    df = standardize_date_column(df)

    df = df.sort_values("年月日")

    return df


# -------------------------------------------------
# 加入欄位前綴
# -------------------------------------------------

def add_prefix(df, file):

    prefix = file.replace(".xlsx","")

    rename_dict = {}

    for col in df.columns:

        if col != "年月日":

            rename_dict[col] = prefix + "_" + col

    df = df.rename(columns=rename_dict)

    return df


# -------------------------------------------------
# 讀取所有資料
# -------------------------------------------------

dfs = []

for file in file_list:

    df = read_excel_file(file)

    df = add_prefix(df, file)

    dfs.append(df)


print("Loaded files:", len(dfs))


# -------------------------------------------------
# 合併資料
# -------------------------------------------------

s2330df = reduce(

    lambda left,right:

    pd.merge(left, right, on='年月日', how='outer'),

    dfs

)


# -------------------------------------------------
# 排序
# -------------------------------------------------

s2330df = s2330df.sort_values("年月日")


# -------------------------------------------------
# 設定 index
# -------------------------------------------------

s2330df = s2330df.set_index("年月日")


# -------------------------------------------------
# 檢查結果
# -------------------------------------------------

print("Merged shape:", s2330df.shape)

print("\nColumns:")

print(s2330df.columns)


print("\nPreview")

display(s2330df.head())


# -------------------------------------------------
# 缺失值檢查
# -------------------------------------------------

missing = s2330df.isnull().sum().sort_values(ascending=False)

print("\nMissing Values")

display(missing.head(20))


# -------------------------------------------------
# 存檔
# -------------------------------------------------

save_path = BASE_PATH + "\\s2330df_merged.csv"

s2330df.to_csv(save_path)

print("Saved:", save_path)
````

````python
# =========================================================
# S2330 台積電多來源資料整併程式
# Google Colab 專用
# 目的：
# 1. 讀取 5 個 Excel 檔
# 2. 跳過前兩列非資料列
# 3. 自動辨識日期欄位（年月日 / 年月 / 日期）
# 4. 統一為 年月日
# 5. 以 年月日 合併成 s2330df
# =========================================================

# -----------------------------
# 1. 掛載 Google Drive
# -----------------------------
# from google.colab import drive
# drive.mount('/content/drive')

# -----------------------------
# 2. 載入套件
# -----------------------------
import os
import re
import warnings
from functools import reduce

import pandas as pd
import numpy as np

warnings.filterwarnings("ignore")

# -----------------------------
# 3. 設定路徑
# -----------------------------
# BASE_PATH = "/content/drive/MyDrive/Colab Notebooks/TEJ2026A/TEJ2026"
# os.chdir(BASE_PATH)

# print("目前工作路徑：", os.getcwd())

# -----------------------------
# 4. 指定檔案清單（5個）
# -----------------------------
file_list = [
    "s2330d20260306.xlsx",
    "三大法人買賣超S2330台積電20260306.xlsx",
    "借券餘額S2330台積電20260306.xlsx",
    "當日沖銷S2330台積電20260306.xlsx",
    "融資融券S2330台積電20260306.xlsx"
]

print("\n=== 檢查檔案是否存在 ===")
for f in file_list:
    full_path = os.path.join(BASE_PATH, f)
    print(f"{f} -> {'存在' if os.path.exists(full_path) else '不存在'}")

# -----------------------------
# 5. 清理欄位名稱
# -----------------------------
def clean_column_names(df):
    """
    清理欄位名稱：
    - 去前後空白
    - 去除換行
    - 去除多餘空白
    """
    df.columns = [
        str(col).strip().replace("\n", "").replace("\r", "").replace(" ", "")
        for col in df.columns
    ]
    return df

# -----------------------------
# 6. 日期欄位標準化
# -----------------------------
def standardize_date_column(df, file_name=""):
    """
    自動辨識日期欄位：
    候選欄位：年月日 / 日期 / 年月
    並統一命名為：年月日
    """

    df = clean_column_names(df)

    date_candidates = ["年月日", "日期", "年月"]
    found_col = None

    for col in date_candidates:
        if col in df.columns:
            found_col = col
            break

    if found_col is None:
        raise ValueError(f"[{file_name}] 找不到日期欄位，欄位如下：{list(df.columns)}")

    # 統一名稱為 年月日
    if found_col != "年月日":
        df = df.rename(columns={found_col: "年月日"})

    # 先轉成字串
    df["年月日"] = df["年月日"].astype(str).str.strip()

    # 處理「年月」可能是 202401 這種格式，補成 20240101
    mask_yyyymm = df["年月日"].str.match(r"^\d{6}$", na=False)
    df.loc[mask_yyyymm, "年月日"] = df.loc[mask_yyyymm, "年月日"] + "01"

    # 如果是 8 碼數字，按 YYYYMMDD 解析
    mask_yyyymmdd = df["年月日"].str.match(r"^\d{8}$", na=False)

    converted = pd.Series(pd.NaT, index=df.index, dtype="datetime64[ns]")

    if mask_yyyymmdd.any():
        converted.loc[mask_yyyymmdd] = pd.to_datetime(
            df.loc[mask_yyyymmdd, "年月日"],
            format="%Y%m%d",
            errors="coerce"
        )

    other_mask = ~mask_yyyymmdd
    if other_mask.any():
        converted.loc[other_mask] = pd.to_datetime(
            df.loc[other_mask, "年月日"],
            errors="coerce"
        )

    df["年月日"] = converted

    before_rows = len(df)
    df = df.dropna(subset=["年月日"]).copy()
    after_rows = len(df)

    print(f"[{file_name}] 日期欄位：{found_col} -> 年月日，刪除無效日期 {before_rows - after_rows} 筆")

    return df

# -----------------------------
# 7. 建立檔名前綴
# -----------------------------
def get_file_prefix(file_name):
    """
    根據檔名建立前綴，避免 merge 後欄位同名衝突
    """
    if "三大法人買賣超" in file_name:
        return "major"
    elif "借券餘額" in file_name:
        return "borrow_balance"
    elif "當日沖銷" in file_name:
        return "day_trade"
    elif "融資融券" in file_name:
        return "margin"
    elif "s2330d" in file_name.lower():
        return "price"
    else:
        base = os.path.splitext(file_name)[0]
        base = re.sub(r"[^A-Za-z0-9\u4e00-\u9fff]+", "_", base)
        return base[:30]

# -----------------------------
# 8. 除日期外加前綴
# -----------------------------
def add_prefix_except_date(df, prefix):
    rename_dict = {}
    for col in df.columns:
        if col != "年月日":
            rename_dict[col] = f"{prefix}_{col}"
    df = df.rename(columns=rename_dict)
    return df

# -----------------------------
# 9. 讀取並整理單一 Excel
# -----------------------------
def load_and_prepare_excel(file_name):
    file_path = os.path.join(BASE_PATH, file_name)
    print(f"\n========== 讀取檔案：{file_name} ==========")

    # 重點：前兩列不是資料
    df = pd.read_excel(file_path, skiprows=2)

    print("原始 shape:", df.shape)

    # 清理欄位名稱
    df = clean_column_names(df)

    # 標準化日期欄位
    df = standardize_date_column(df, file_name=file_name)

    # 刪除完全重複列
    before_dup = len(df)
    df = df.drop_duplicates()
    after_dup = len(df)
    print(f"[{file_name}] 刪除完全重複列：{before_dup - after_dup} 筆")

    # 日期排序
    df = df.sort_values("年月日").copy()

    # 若同一日期重複，保留最後一筆
    dup_date_count = df.duplicated(subset=["年月日"]).sum()
    if dup_date_count > 0:
        print(f"[{file_name}] 發現重複日期 {dup_date_count} 筆，保留最後一筆")
        df = df.drop_duplicates(subset=["年月日"], keep="last")

    # 欄位加前綴
    prefix = get_file_prefix(file_name)
    df = add_prefix_except_date(df, prefix)

    print(f"[{file_name}] 整理後 shape:", df.shape)
    print(f"[{file_name}] 日期範圍: {df['年月日'].min()} ~ {df['年月日'].max()}")

    return df

# -----------------------------
# 10. 逐一讀取 5 個檔案
# -----------------------------
df_list = []

for file_name in file_list:
    df_temp = load_and_prepare_excel(file_name)
    df_list.append(df_temp)

print("\n成功載入資料表數量：", len(df_list))

# -----------------------------
# 11. 以 年月日 做 outer merge
# -----------------------------
print("\n========== 開始合併 ==========")

s2330df = reduce(
    lambda left, right: pd.merge(left, right, on="年月日", how="outer"),
    df_list
)

# -----------------------------
# 12. 合併後整理
# -----------------------------
s2330df = s2330df.sort_values("年月日").reset_index(drop=True)

print("合併後 shape:", s2330df.shape)

# -----------------------------
# 13. 缺失值檢查
# -----------------------------
missing_df = pd.DataFrame({
    "欄位名稱": s2330df.columns,
    "缺失值筆數": s2330df.isnull().sum().values,
    "缺失比例(%)": (s2330df.isnull().sum().values / len(s2330df) * 100).round(2)
}).sort_values("缺失值筆數", ascending=False)

print("\n========== 缺失值檢查（前20欄）==========")
display(missing_df.head(20))

# -----------------------------
# 14. 數值欄位嘗試轉型
# -----------------------------
for col in s2330df.columns:
    if col != "年月日":
        s2330df[col] = pd.to_numeric(s2330df[col], errors="ignore")

print("\n========== 資料型態 ==========")
print(s2330df.dtypes)

# -----------------------------
# 15. 設定 年月日 為 index
# -----------------------------
s2330df["年月日"] = pd.to_datetime(s2330df["年月日"])
s2330df = s2330df.sort_values("年月日").set_index("年月日")

print("\n========== 最終 s2330df ==========")
print("shape:", s2330df.shape)
display(s2330df.head())

# -----------------------------
# 16. 存檔
# -----------------------------
csv_path = os.path.join(BASE_PATH, "s2330df_merged.csv")
xlsx_path = os.path.join(BASE_PATH, "s2330df_merged.xlsx")

s2330df.to_csv(csv_path, encoding="utf-8-sig")
s2330df.to_excel(xlsx_path)

print("\n已存檔：")
print(csv_path)
print(xlsx_path)

# -----------------------------
# 17. 若要保留 年月日 為一般欄位
# -----------------------------
s2330df_reset = s2330df.reset_index()

print("\n可另外使用 s2330df_reset（年月日為一般欄位）")
display(s2330df_reset.head())
````

````python
# =========================================================
# TSMC 量化交易分析範例 (Colab)
# =========================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 假設你已經有 s2330df_reset
df = s2330df_reset.copy()

# ---------------------------------------------------------
# 1️⃣ 先做缺失值處理
# ---------------------------------------------------------
df.fillna(method='ffill', inplace=True)  # 向前填補
df.fillna(method='bfill', inplace=True)  # 向後填補

# ---------------------------------------------------------
# 2️⃣ 計算技術指標
# ---------------------------------------------------------

# 移動平均線
df['MA5'] = df['price_收盤價(元)'].rolling(5).mean()
df['MA20'] = df['price_收盤價(元)'].rolling(20).mean()
df['MA60'] = df['price_收盤價(元)'].rolling(60).mean()

# 日報酬率與累積報酬
df['daily_return'] = df['price_收盤價(元)'].pct_change()
df['cum_return'] = (1 + df['daily_return']).cumprod()

# RSI 指標計算
delta = df['price_收盤價(元)'].diff()
up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)
roll_up = up.rolling(14).mean()
roll_down = down.rolling(14).mean()
RS = roll_up / roll_down
df['RSI'] = 100 - (100 / (1 + RS))

# ---------------------------------------------------------
# 3️⃣ 三大法人資金流向
# ---------------------------------------------------------
# 外資 + 投信 + 自營買超 (單位: 千股)
df['total_major_buy'] = df['major_外資買賣超(千股)'] + df['major_投信買賣超(千股)'] + df['major_自營買賣超(千股)']

# 資金流向訊號
df['major_signal'] = np.where(df['total_major_buy'] > 0, 1, -1)

# ---------------------------------------------------------
# 4️⃣ 交易策略訊號
# ---------------------------------------------------------
# 技術訊號: MA5 與 MA20 黃金交叉/死亡交叉
df['ma_signal'] = 0
df.loc[df['MA5'] > df['MA20'], 'ma_signal'] = 1  # 多頭
df.loc[df['MA5'] < df['MA20'], 'ma_signal'] = -1  # 空頭

# 結合法人與技術訊號
df['final_signal'] = 0
df.loc[(df['ma_signal'] == 1) & (df['major_signal'] == 1), 'final_signal'] = 1  # 買進
df.loc[(df['ma_signal'] == -1) & (df['major_signal'] == -1), 'final_signal'] = -1  # 賣出

# ---------------------------------------------------------
# 5️⃣ 回測策略簡單績效
# ---------------------------------------------------------
df['strategy_return'] = df['daily_return'] * df['final_signal'].shift(1)  # 隔日買賣
df['cum_strategy_return'] = (1 + df['strategy_return']).cumprod()

# ---------------------------------------------------------
# 6️⃣ 繪圖
# ---------------------------------------------------------
plt.figure(figsize=(16,8))
plt.plot(df['年月日'], df['cum_return'], label='持有股票累積報酬')
plt.plot(df['年月日'], df['cum_strategy_return'], label='策略累積報酬')
plt.title('台積電量化交易策略績效')
plt.xlabel('日期')
plt.ylabel('累積報酬')
plt.legend()
plt.show()

# ---------------------------------------------------------
# 7️⃣ 最後交易建議
# ---------------------------------------------------------
latest_signal = df['final_signal'].iloc[-1]
if latest_signal == 1:
    print("📈 建議：近期可考慮買進台積電")
elif latest_signal == -1:
    print("📉 建議：近期可考慮賣出台積電或空手觀望")
else:
    print("⚪ 建議：暫時觀望，無明確訊號")
````
