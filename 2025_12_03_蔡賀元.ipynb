```python
# E3-直角三角形面積(函數)-0.py


def triangle(b,h):
  """
  # def 定義一個triangle方法
  # 傳入b和h兩個變數參數
  """
  # 運算式:0.5*b*h
  # 使用return:傳回運算式的答案
  return 0.5*b*h

# 定義兩個變數
# 用:逗號分隔同時定義
# b=3 ; h=4
# # a,b=1,2

b=3
h=4

# 呼叫方法 並且顯示
print("三角函數的面積:",triangle(b,h) )
```

```python
# E3-海龍公式計算三角形面積.py
# 海龍公式
a=3; b=6; c=4;
# a=input("# a ")


# 運算式 存入變數s
# (...) 優先計算
s= (a+b+c)/2
# * 乘
# ** 次方
area= (s*(s-a)*(s-b)*(s-c))**0.5
print("面積= %0.2f"% area)
```
